```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
file = os.listdir('data')
print(file)
```

```python
labels = []
faces_paths = []

for f in file:
    images = os.listdir('data/'+f)
    for pic in images:
        faces_paths.append('data/'+f+'/'+pic)
        labels.append(f)
```

```python
print(labels)
# print(faces_paths)
```

```python
imgs = []

for fx in faces_paths:
    img = cv2.imread(fx)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    imgs.append(img)
```

```python
print(imgs)
```

```python
all_faces = []

for i in range(12):
    im = imgs[i][20:400,50:800,:]
    img = cv2.resize(im, (100,100))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    all_faces.append(img)
    plt.imshow(img, cmap='gray')
    plt.show()
    
for i in range(12,28):
    im = imgs[i][80:500,500:900,:]
    img = cv2.resize(im, (100,100))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    all_faces.append(img)
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
all_faces = np.asarray(all_faces)
```

```python
lab = np.asarray(labels[:])

np.save('facelabels',lab)
np.save('facedata', all_faces)
```
