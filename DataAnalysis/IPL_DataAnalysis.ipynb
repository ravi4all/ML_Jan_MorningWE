```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ipl/matches.csv')
```

# Matches Played Per Season

```python
sns.countplot(df['season'], data=df)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(df['winner'], data=df)
plt.xticks(rotation=45)
plt.title("Number of matches won by a team")
plt.show()
```

```python
df.drop_duplicates(subset='season', keep='last')[['season','winner']].reset_index(drop=True)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(df['venue'], data=df)
plt.xticks(rotation='vertical')
plt.show()
```
